# Data Visualization with Matplotlib library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Basic Visualization

```python
height = [150,160,165,185]
weight = [70,80,90,100]
plt.plot(height,weight)
```

## Judul dan label

```python
plt.plot(height,weight)
# to give title to the graph/figure
plt.title('Height vs Weight graph')
# Add x label
plt.xlabel("height")
# Add y Label
plt.ylabel("weight")
```

## Plotting lebih dari 1 garis

```python
calories_burnt = [65,75,95,99]
#Plotting 2 lines
plt.plot(height, calories_burnt)
plt.plot(height, weight)

# to give title to the graph/figure
plt.title('Calorie vs Weight over the Height graph')

# Add x label
plt.xlabel("height")

# Add y Label
plt.ylabel("weight and calorie")
```

## penambahan legend ke plot/graph/figure

```python
#Plotting 2 lines
plt.plot(height, calories_burnt)
plt.plot(height, weight)

# to give title to the graph/figure
plt.title('Calorie vs Weight over the Height graph')

# Add x label
plt.xlabel("height")

# Add y Label
plt.ylabel("weight and calorie")

# Adding legend to the plot/graph/figure
plt.legend(labels=['Calories',"weight"],loc='lower right')
```

## Figure size

```python
# figure size in inches, should be at the beginning
plt.figure(figsize=(13,7))

#Plotting 2 lines
plt.plot(height, calories_burnt)
plt.plot(height, weight)

# to give title to the graph/figure
plt.title('Calorie vs Weight over the Height graph')

# Add x label
plt.xlabel("height")

# Add y Label
plt.ylabel("weight and calorie")

# Adding legend to the plot/graph/figure
plt.legend(labels=['Calories',"weight"],loc='lower right')
```

## Mengganti bentuk garis

```python
# figure size in inches, should be at the beginning
plt.figure(figsize=(13,5))

#Plotting 2 lines
plt.plot(height, calories_burnt, '--o')
plt.plot(height, weight,  'r-o')

# to give title to the graph/figure
plt.title('Calorie vs Weight over the Height graph')

# Add x label
plt.xlabel("height")

# Add y Label
plt.ylabel("weight and calorie")

# Adding legend to the plot/graph/figure
plt.legend(labels=['Calories',"weight"],loc='lower right')
```

## Plotting data from dataframe
### Plotting by class
Misalkan suatu perusahaan pesawat terbang ingin tahu perbedaan rata-rata usia penumpang tiap kelas

```python
passenger_data = pd.read_csv('https://ngodingdata.com/wp-content/uploads/2020/05/titanic.csv')
passenger_data
```

```python
#Menghitung rata-rata usia per kelas
age_by_pclass = passenger_data.groupby('Pclass')['Age'].mean()
age_by_pclass
```

```python
x = age_by_pclass.index.tolist()      #x-axis is Passenger class (Pclass)
y = age_by_pclass.values.tolist()     #y-axis rata-rata usia
```

```python
#Make a bar chart
plt.bar(x,y,color=['red','orange','pink'])
```

Kita bisa menjelaskan ke perusahaan bahwa fasilitas kelas 1 lebih diminati penumpang berusia sekitar 36 tahun
